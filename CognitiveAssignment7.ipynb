Part I: Randomized Sales Data GeneraƟon (NumPy)
1. IniƟalize Random Seed
2. Generate Sales Data
 Create a NumPy array of random sales data for 12 months (shape = (12, 4)),
where each column represents a product category:
 Electronics, Clothing, Home & Kitchen, Sports.
 Sales values should be between 1000 and 5000 units per month.
 The range should be randomized using the seed.
3. Convert to DataFrame
 Convert the cleaned NumPy array into a Pandas DataFrame with:
i. Columns: Electronics, Clothing, Home & Kitchen, Sports
ii. Index: Months from Jan to Dec
Part II: Data ManipulaƟon and Analysis (Pandas)
1. Display the first 5 rows and summary staƟsƟcs of the DataFrame.
2. Calculate the total sales per category and total sales per month.
3. Calculate the average sales growth between consecuƟve months for each category.
4. Add the following columns to the DataFrame:
 Total Sales: Sum of all categories per month.
 Growth Rate: Percentage change in Total Sales from the previous month.
5. Apply a randomized discount:
 If your roll number is even, apply a 10% discount to the Electronics category.

 If odd, apply a 15% discount to the Clothing category.

Part III: VisualizaƟons (Matplotlib and Seaborn)
1. Plot monthly sales trends for each category using line plots.
2. Create the following plots:
 Box plots to show the sales distribuƟon for each category.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set random seed (Replace with your roll number)
roll_number = int(input("Enter your roll number: "))
np.random.seed(roll_number)

# Generate random sales data for 12 months (4 categories)
sales_data = np.random.randint(1000, 5000, size=(12, 4))

# Convert to DataFrame
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
categories = ['Electronics', 'Clothing', 'Home & Kitchen', 'Sports']
df = pd.DataFrame(sales_data, columns=categories, index=months)

# Display first 5 rows
display(df.head(5))

# Summary statistics
display(df.describe())

# Total sales per category and per month
df['Total Sales'] = df.sum(axis=1)
category_sales = df[categories].sum()

display(category_sales)
display(df)

# Calculate monthly growth rate
df['Growth Rate'] = df['Total Sales'].pct_change() * 100

# Apply discount based on roll number
if roll_number % 2 == 0:
    df['Electronics'] *= 0.9  # 10% discount
else:
    df['Clothing'] *= 0.85  # 15% discount

# Visualization
plt.figure(figsize=(12, 6))
for category in categories:
    plt.plot(df.index, df[category], marker='o', label=category)
plt.title("Monthly Sales Trends")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.legend()
plt.show()

# Box plot for sales distribution
plt.figure(figsize=(10, 6))
sns.boxplot(data=df[categories])
plt.title("Sales Distribution by Category")
plt.show()
```

each row

```python
import numpy as np

# Given array
array = np.array([[1, -2, 3], [-4, 5, -6]])

# (i) Element-wise absolute value
abs_array = np.abs(array)

# (ii) Percentiles
flat_percentiles = np.percentile(array, [25, 50, 75])  # Flattened array
col_percentiles = np.percentile(array, [25, 50, 75], axis=0)  # Column-wise
row_percentiles = np.percentile(array, [25, 50, 75], axis=1)  # Row-wise

# (iii) Mean, Median, Standard Deviation
flat_mean = np.mean(array)
flat_median = np.median(array)
flat_std = np.std(array)

col_mean = np.mean(array, axis=0)
col_median = np.median(array, axis=0)
col_std = np.std(array, axis=0)

row_mean = np.mean(array, axis=1)
row_median = np.median(array, axis=1)
row_std = np.std(array, axis=1)

# Print results
print("Element-wise Absolute Value:\n", abs_array, "\n")

print(f"Flattened Array Percentiles (25th, 50th, 75th):\n{flat_percentiles}\n")
print(f"Column-wise Percentiles:\n{col_percentiles}\n")
print(f"Row-wise Percentiles:\n{row_percentiles}\n")

print(f"Flattened Array: Mean={flat_mean}, Median={flat_median}, Std Dev={flat_std}\n")
print(f"Column-wise: Mean={col_mean}, Median={col_median}, Std Dev={col_std}\n")
print(f"Row-wise: Mean={row_mean}, Median={row_median}, Std Dev={row_std}")
```

Q.4 WAP to swap two elements in a list in Python Using a Temporary Variable.

```python
#swap two elements in a list
def swap_elements(lst, pos1, pos2):
    if pos1 < 0 or pos2 < 0 or pos1 >= len(lst) or pos2 >= len(lst):
        print("Invalid positions!")
        return lst

    # Swapping using a temporary variable
    temp = lst[pos1]
    lst[pos1] = lst[pos2]
    lst[pos2] = temp

    return lst

my_list = [10, 20, 30, 40, 50]
pos1, pos2 = 1, 3  # Swap elements at index 1 and 3

print("Original List:", my_list)
swapped_list = swap_elements(my_list, pos1, pos2)
print("List after swapping:", swapped_list)

```

Q5: Write a python program to swap two elements in a set y conerting it into list first

```python
from collections import OrderedDict

def swap_elements_ordered_set(my_set, pos1, pos2):
    # Convert set to an ordered list while maintaining uniqueness
    ordered_list = list(OrderedDict.fromkeys(my_set))

    # Validate positions
    if pos1 < 0 or pos2 < 0 or pos1 >= len(ordered_list) or pos2 >= len(ordered_list):
        print("Invalid positions!")
        return my_set  # Return original set

    # Swap elements
    ordered_list[pos1], ordered_list[pos2] = ordered_list[pos2], ordered_list[pos1]

    # Convert back to an ordered set using OrderedDict
    return OrderedDict.fromkeys(ordered_list)  # Preserves order

my_set = {10, 20, 30, 40, 50}  # Maintains order in Python 3.7+
pos1, pos2 = 1, 3  # Swap elements at index 1 and 3

print("Original set:", my_set)
swapped_set = swap_elements_ordered_set(my_set, pos1, pos2)
print("Ordered set after swapping:", list(swapped_set.keys()))  # Display ordered set
```
