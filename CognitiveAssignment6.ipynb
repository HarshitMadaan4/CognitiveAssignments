Q1. Ask the user to enter a value (e.g., M) for a mathematical function.
 Generate x values from -10 to 10 using np.linspace().
 Compute y values for:
o y= M⋅x
2
o y=M⋅sin(x)
 Plot both functions in a single figure:
o Use different colors and line styles.
o Add legend, grid, and title.

```python
import numpy as np
import matplotlib.pyplot as plt

M= float(input('enter value of M: '))
x= np.linspace(-10,10,100)
y1= M*x**2
y2= M*np.sin(x)

plt.plot(x,y1,'r--',label='y=M*x^2')
plt.plot(x,y2,'b-',label='y=M*sin(x)')
plt.legend()
plt.grid()
plt.title('mathematical function')
plt.show()
```

Q2:Create a dataset of five subjects and scores.
 Convert it into a Pandas DataFrame.
 Plot the scores using a Seaborn bar plot with:
o Different colors for each bar.
o Annotations on top of each bar.
o Title, axis labels, and grid.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

dataset = {
    'Subjects': ['Math', 'Science', 'History', 'English', 'Computer Science'],
    'Scores': [85, 90, 78, 88, 95]
}

dataframe = pd.DataFrame(dataset)

color_palette = sns.color_palette('pastel')

plt.figure(figsize=(8, 5))
ax = sns.barplot(x='Subjects', y='Scores', data=dataframe, palette=color_palette)

for i, score in enumerate(dataframe['Scores']):
    ax.text(i, score + 1, str(score), ha='center', fontsize=12, fontweight='bold')

plt.title("Student Scores in Different Subjects", fontsize=14, fontweight='bold')
plt.xlabel("Subjects", fontsize=12)
plt.ylabel("Scores", fontsize=12)

plt.xticks(rotation=30)

plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.show()
```

Q3. Write a Python script to select your roll number as seed for NumPy and
then generate a dataset of 50 values using np.random.randn().
 Create a 2x2 subplot layout displaying:
o Line plot showing cumulative sum.
o Scatter plot with random noise.
 Customize with titles, axis labels, and grids.

```python
import numpy as np
import matplotlib.pyplot as plt


roll_number = int(input("Enter your roll number: "))
np.random.seed(roll_number)

# Generate dataset of 50 values using np.random.randn()
data = np.random.randn(50)

# Create a 2x2 subplot layout
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# Line plot showing cumulative sum
axes[0, 0].plot(np.cumsum(data), marker='o', linestyle='-')
axes[0, 0].set_title("Cumulative Sum Line Plot")
axes[0, 0].set_xlabel("Index")
axes[0, 0].set_ylabel("Cumulative Sum")
axes[0, 0].grid(True)

# Scatter plot with random noise
noise = np.random.randn(50)
axes[0, 1].scatter(range(50), data + noise, color='r', alpha=0.6)
axes[0, 1].set_title("Scatter Plot with Random Noise")
axes[0, 1].set_xlabel("Index")
axes[0, 1].set_ylabel("Value with Noise")
axes[0, 1].grid(True)

# Hide the unused subplots
axes[1, 0].axis("off")
axes[1, 1].axis("off")

# Adjust layout and show the plots
plt.tight_layout()
plt.show()
```

Q.4 Download Data-set from the below link
https://github.com/AnjulaMehto/MCA/blob/main/company_sales_data.csv
Apply ‘seaborn’ library to do the following.
1. Read Total profit of all months and show it using a line plot.
2. Read all product sales data and show it using a multiline plot.
3. Plot bar chart for all the features/attributes.

```python
from google.colab import files
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Upload file
uploaded = files.upload()

# Read the uploaded CSV file
filename = list(uploaded.keys())[0]  # Get the uploaded file name
df = pd.read_csv(filename)  # Read it into a DataFrame

# Line plot for total profit over months
plt.figure(figsize=(10, 5))
sns.lineplot(data=df, x='month_number', y='total_profit', marker='o', color='b')
plt.title('Total Profit Over Months')
plt.xlabel('Month')
plt.ylabel('Total Profit')
plt.grid()
plt.show()

# Multi-line plot for product sales
plt.figure(figsize=(10, 5))
for column in df.columns[1:-1]:  # Excluding 'month_number' and 'total_profit'
    sns.lineplot(data=df, x='month_number', y=column, label=column)
plt.title('Product Sales Over Months')
plt.xlabel('Month')
plt.ylabel('Units Sold')
plt.legend()
plt.grid()
plt.show()

# Bar chart for all features
plt.figure(figsize=(12, 6))
df.set_index('month_number').plot(kind='bar', figsize=(12, 6))
plt.title('Monthly Sales Data')
plt.xlabel('Month')
plt.ylabel('Values')
plt.legend()
plt.grid()
plt.show()
```
